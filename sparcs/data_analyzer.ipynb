```python
import numpy as np
import cv2 as cv
from helper_functions import *
from time import time, sleep
DIST_AHEAD = 0.3

imgs = np.load('saved_tests/imgs.npy')
locs = np.load('saved_tests/locs.npy')
path = np.load('sparcs_path.npy').T



print(f'imgs.shape = {imgs.shape}')
print(f'locs.shape = {locs.shape}')

cv.namedWindow('Frame', cv.WINDOW_NORMAL)

for img, loc in zip(imgs, locs):
    x = loc[0]
    y = loc[1]
    yaw = loc[2]

    #get heading error
    print(f'x: {x} \ny: {y} \nyaw: {np.rad2deg(yaw)}\n')
    p = np.array([x,y]).T #current position of the car
    min_index = np.argmin(np.linalg.norm( path-p,axis=1)) #index of clostest point on path
    pa =  path[(min_index + int(100*DIST_AHEAD)) % len( path)] #point ahead
    yaw_ref = np.arctan2(pa[1]-p[1],pa[0]-p[0]) #yaw reference in world frame
    he = diff_angle(yaw_ref, yaw) #heading error
    # print(f'x_a: {pa[0]} \ny_a: {pa[1]} \nhe: {np.rad2deg(he)}\n')

    img = draw_angle(img, he)

    cv.imshow('Frame', img)
    cv.waitKey(1)

    sleep(0.030)



cv.destroyAllWindows()
```
